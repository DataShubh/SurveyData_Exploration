```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel("SurveyMonkey_Data.xlsx", sheet_name= "UpdatedData" )
```

```python
dataset
```

```python
data = dataset.copy()
```

```python
data.columns
```

```python
cols=['Start Date', 'End Date',
       'Email Address', 'First Name', 'Last Name', 'Custom Data']
```

```python
data= data.drop(cols, axis =1) 
```

```python
data.columns
```

```python
list(data.columns)[:8]
```

```python
id_vars= list(data.columns)[:8]
value_vars = list(data.columns)[8:]
```

```python
data.melt(id_vars= id_vars, value_vars= value_vars, var_name = "Question + Subquestion", value_name = "Response")
```

```python
data_melted = data.melt(id_vars= id_vars, value_vars= value_vars, var_name = "Question + Subquestion", value_name = "Response")
```

```python
data_melted['Identify which division you work in.-Response'].unique()
```

```python
question_format = pd.read_excel("SurveyMonkey_Data.xlsx", sheet_name= "Question")
```

```python
question_format
```

```python
question_format.drop(columns =[ "Raw Question", "Raw Subquestion", "Subquestion"], axis = 1, inplace= True)
```

```python
data_merged = pd.merge(left= data_melted, right=question_format, how="left", left_on= "Question + Subquestion", right_on="Question + Subquestion")
print("Orginal_dataset shape", data_melted.shape)
print("merged_dataset shape", data_merged.shape)
data_merged
```

```python
data_melted.shape
```

```python
data_melted["Respondent ID"].unique().shape
```

```python
data_response = data_melted.groupby("Question + Subquestion")["Response"].count().reset_index()
```

```python
data_response.rename(columns = {"Response" : "TotalRespondant"}, inplace = True)
```

```python
data_response
```

```python
data_merged2 = pd.merge(left= data_merged, right=data_response, how="left", left_on= "Question + Subquestion", right_on="Question + Subquestion")
print("Orginal_dataset shape", data_merged.shape)
print("merged_dataset shape", data_merged2.shape)
data_merged2
```

```python
data_response2 =data_melted.groupby(["Question + Subquestion", "Response"])[ "Respondent ID"].count().reset_index()
data_response2.rename(columns = {"Respondent ID" : "Same Answer"}, inplace = True)
```

```python
data_merged3 = pd.merge(left= data_merged2, right=data_response2, how="left", left_on= ["Question + Subquestion", "Response"], right_on=["Question + Subquestion", "Response"])
print("Orginal_dataset shape", data_merged2.shape)
print("merged_dataset shape", data_merged3.shape)
data_merged3
```

```python
data_merged3.columns
```

```python
data_merged3.rename(columns = {"Identify which division you work in.-Response": "Primary Division", "Identify which division you work in.-Other (please specify)": "Secondary Divison",
                              "Which of the following best describes your position level?-Response":"Position", "Which generation are you apart of?-Response": 'Generation',
                              'Please select the gender in which you identify.-Response': "Gender",'Which duration range best aligns with your tenure at your company?-Response': "Tenure",
                              "Which of the following best describes your employment type?-Response": "Employment Type"}, inplace = True)
```

```python
data_melted.groupby(["Question + Subquestion", "Response", "Which of the following best describes your position level?-Response"])[ "Respondent ID"].count().reset_index()
```
